```python

import os
import re
import pandas as pd
import warnings
from pathlib import Path

warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")

# Optional: keep filenames safe across OSes
INVALID_FILENAME_REGEX = re.compile(r'[<>:"/\\|?*\x00-\x1F]')

def sanitize_filename_part(s: str) -> str:
    s = INVALID_FILENAME_REGEX.sub("_", s.strip())
    s = re.sub(r"\s+", " ", s).strip(" .")
    return s or "Unnamed"

def rename_files_by_contract_id(folder_path):
    folder = Path(folder_path)
    for filename in os.listdir(folder):
        if not filename.lower().endswith(".xlsx"):
            continue

        file_path = folder / filename

        try:
            # Read the Excel file
            df = pd.read_excel(file_path, engine="openpyxl")

            # Normalize headers (remove line breaks and trim)
            df.columns = [str(col).replace("\n", "").strip() for col in df.columns]

            # Get Contract value from first row (col name "Contract")
            contract_val = df.loc[0, "Contract"]  # adjust if it's "Contract ID"
            contract_id = "" if pd.isna(contract_val) else str(contract_val).strip()

            if not contract_id or contract_id.lower() == "nan":
                print(f"Skipping {filename}: Contract value missing")
                continue

            # Build target name and path
            safe_id = sanitize_filename_part(contract_id)
            new_filename = f"Contract ID - {safe_id}.xlsx"
            new_path = folder / new_filename

            # If source and target are identical, skip
            if file_path.name == new_filename:
                print(f"Already named correctly: {filename}")
                continue

            # Replace if exists (atomic on most OSes)
            os.replace(file_path, new_path)
            print(f"Renamed (replaced if existed): {filename} → {new_filename}")

        except Exception as e:
            print(f"Error processing {filename}: {e}")

if __name__ == "__main__":
    folder = r"C:\Users\IlaBarshilia\ACME Barricades\FDOT Web Scraping - FDOT Web Scraping Data\Source Data\fdot_downloads_2026-Jan-23"
    rename_files_by_contract_id(folder)
```
