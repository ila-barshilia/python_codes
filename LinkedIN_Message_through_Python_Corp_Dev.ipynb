```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pyperclip
import time

# --- Configuration ---
USERNAME = "ilabarshilia@gmail.com"
PASSWORD = "Iambrilliant20"
PROFILE_URL = "https://www.linkedin.com/in/andrew-sterling-hoosier/"
MESSAGE_TEXT = "Hi Andrew, I am conducting this trial from Python Code, please bear with me. Sent from Python code"

# --- Setup Stealth Chrome ---
options = uc.ChromeOptions()
options.add_argument("user-data-dir=C:/Users/IlaBarshilia/AppData/Local/Google/Chrome/User Data")
options.add_argument("--disable-blink-features=AutomationControlled")
driver = uc.Chrome(options=options)
wait = WebDriverWait(driver, 15)

try:
    print("🔐 Logging into LinkedIn...")
    driver.get("https://www.linkedin.com/login")
    wait.until(EC.presence_of_element_located((By.ID, "username"))).send_keys(USERNAME)
    driver.find_element(By.ID, "password").send_keys(PASSWORD)
    driver.find_element(By.XPATH, "//button[@type='submit']").click()

    wait.until(EC.presence_of_element_located((By.ID, "global-nav-search")))
    time.sleep(5)
    print("✅ Logged in successfully.")

    print(f"🌐 Navigating to profile: {PROFILE_URL}")
    driver.get(PROFILE_URL)
    time.sleep(5)

    print("🔍 Searching for Message button...")
    message_button = wait.until(EC.element_to_be_clickable(
        (By.XPATH, "//button[contains(text(), 'Message')]")
    ))
    print("✅ Message button found. Clicking...")
    message_button.click()
    time.sleep(3)

    print("✍️ Pasting message...")
    pyperclip.copy(MESSAGE_TEXT)
    message_box = wait.until(EC.presence_of_element_located(
        (By.XPATH, "//div[contains(@class, 'msg-form__contenteditable')]")
    ))
    message_box.click()
    message_box.send_keys(Keys.CONTROL, 'v')
    time.sleep(1)

    print("✅ Message pasted. You can now click Send manually.")

except Exception as e:
    print("❌ Error occurred:", e)
    print("📄 Page title:", driver.title)
    print("🔎 First 1000 characters of page source:")
    print(driver.page_source[:1000])
```
